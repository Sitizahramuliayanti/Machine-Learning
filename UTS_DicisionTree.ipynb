### Preprocessing Data

```python
# Import library sklearn.preprocessing menyediakan berbagai alat untuk memproses preprocessing dan menyiapkan data sebelum data tersebut digunakan untuk melatih model machine learning
from sklearn.preprocessing import LabelEncoder
```

```python
#Mengubah values kolom yang bertipe data string/objek manjadi numerik
for col in ['gender', 'smoking_history']:
    label_encoder = LabelEncoder()
    dataset[col] = label_encoder.fit_transform(dataset[col])
```

####  Dataset yang memiliki tipe data string (gender dan smoking_history) diubah menjadi numerik  setelah diubah, dataset akan memiliki nilai yang terkode numerik semua, yang bisa digunakan lebih lanjut dalam proses analisis atau pemodelan machine learning. Misalnya, kolom 'gender' sebelumnya berisi 'male' dan 'female', setelah transformasi, diubah sebagai 0 dan 1. Ini memungkinkan algoritma machine learning untuk menggunakan informasi ini dalam analisis. 

```python
# Menampilkan dataset yang berisi data numerik semua
dataset
```

```python
# Memisahkan kolom target (diabetes) dari dataset
# axis 1 (menghapus vertikal) 0 (menghapus horizontal)
x = dataset.drop('diabetes', axis = 1)
y = dataset['diabetes']
```

#### Karena kolom "diabetes" akan dijadikan sebagai target, maka harus di keluarkan dari dataset sehingga dataset yang baru akan di simpan pada x. dan kolom target akan di simpan pada y

```python
# Menampilkan dataset yang baru
x
```

```python
# Menampilkan kolom target (diabetes)
y
```

### Split Data

```python
# Splitting data training dan testing
from sklearn.model_selection import train_test_split
```

```python
# 70% : 30%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

```python
x_train.shape
```

```python
y_train.shape
```

### Modelling

```python
from sklearn.tree import DecisionTreeClassifier
modelDT = DecisionTreeClassifier(max_depth = 5)
modelDT.fit(x_train, y_train)
```

```python
hasil_prediksi_dt = modelDT.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Akurasi Model: ', accuracy_score(y_test, hasil_prediksi_dt) * 100)
```

```python
# visualisasi model decision tree
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize = (40, 20))
plot_tree(modelDT, fontsize = 24)
plt.show()
```
